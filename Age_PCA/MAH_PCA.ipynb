# "Cleaning" MAH data by adding lower mass values to small mahs

```python
fed = []
def clean_mah(mahs, data, mah_length = 70, plim=1000):
    cleaned = np.zeros((len(mahs), mah_length))
    nhalos = len(mahs)
    for i in range(nhalos):
        mah = mahs[i]
        m = mah[0]
        l = len(mah)
        if l>mah_length-1:
            cleaned[i,:] = mah[:mah_length]
        else:
            gam, bet = data['McBride gamma'][i], data['McBride beta'][i]
            if (gam > -plim)*(gam < plim)*(bet > -plim)*(bet < plim):
                tocomplete = m*mcbride_mah(reds[l:mah_length], gam, bet)
                cleaned[i, l:mah_length] = tocomplete
            else:
                cleaned[i, l:mah_length] = 1e12
            cleaned[i, :l] = mah
            
    return cleaned

def clean2(mahs, data, mah_length = 80, mlim=0):
    cleaned = []
    nhalos = len(mahs)
    for i in range(nhalos):
        mah = mahs[i]
        m = mah[0]
        if m>mlim:
            l = len(mah)
            if l>mah_length-1:
                cleaned.append(mah[:mah_length]/m)
    return np.array(cleaned)
c2 = clean2(mahs, zdat, mlim=1e13)
#cleaned = clean_mah(mahs, data)
```
